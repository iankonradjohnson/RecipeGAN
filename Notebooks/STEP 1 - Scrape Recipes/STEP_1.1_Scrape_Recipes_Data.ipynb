```
import pandas as pd
import ast
from tqdm import notebook
import allrecipes
import _abstract
import _utils
from scrape_me import scrape_me
import concurrent.futures
import time
from requests.exceptions import ConnectionError
import random
```

```
R_DF = pd.read_csv('/content/drive/My Drive/Data Science Projects/RecipeApp/data/myData/ALL_RECIPES.csv', index_col=0)
```

```
R_DF.head()
```

```
li = []

for index, row in notebook.tqdm(R_DF.iterrows(), total=len(R_DF)):
  for recipe in ast.literal_eval(row[0]):
    li.append({'recipe_name': recipe, 'category': row[1]})
    
```

```
df = pd.DataFrame(li)
```

```
df.drop_duplicates(subset ="recipe_name", 
                     keep = 'first',
                     inplace = True) 
```

```
def scrape_recipe(url):
  dict = {}
  try:
    scraper = scrape_me(url)
    # time.sleep(random.randrange(1,4))
    dict = {
            "title": scraper.title(),
            "rating": scraper.ratings(),
            "total_time": scraper.total_time(),
            "yeilds": scraper.yields(),
            "ingredients": scraper.ingredients(),
            "instructions": scraper.instructions(),
            "category": scraper.category()
            }
    return dict
  except ConnectionError as e:
    print(e)
    time.sleep(random.randrange(40,60))
  except:
    return {}
  return dict
```

```
recipes = df['recipe_name'].to_list()
```

```
threads = min(40, len(recipes))
with concurrent.futures.ThreadPoolExecutor(max_workers=threads) as executor:
    r = list(notebook.tqdm(executor.map(scrape_recipe, recipes), total=len(recipes)))

```

```
pd.DataFrame(r).to_csv('/content/drive/My Drive/Data Science Projects/RecipeApp/data/myData/ALL_RECIPES_FINAL.csv')
```
